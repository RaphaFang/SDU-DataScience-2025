```python
countries_list = list(df['name'])
countries_life_dict = {}

for n in countries_list:
    countries_life_dict[n] = [float(i) for i in list(df.loc[df['name'] == n].iloc[0])[2::]]

print(len(countries_life_dict))
```

```python
import matplotlib.pyplot as plt

years = list(df.columns)[2::]
for country, values in countries_life_dict.items():
    plt.plot(years, values, label=country)

plt.xlabel("Year")
plt.ylabel("Value")
plt.title("life expected")

# plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
# plt.tight_layout()
# plt.figure(figsize=(1000, 10000))

plt.show()
```

```python
# absolute growth
df_diff = df.drop(columns=['name','geo']).T.diff()
df_diff
```

```python
df_diff.plot(legend=False,
            xlabel="years",
            ylabel="absolute growth rate",
            title="absolute growth")
```

```python
df_diff.index = df_diff.index.astype(int)
df_diff.loc[1900:1950, 1].plot()
```

```python
# relative growth
df_pct_change = df.drop(columns=['name','geo']).T.pct_change()
df_pct_change
```

```python
df_pct_change.plot(legend=False,
                   xlabel="years",
                   ylabel="relative growth percentage",
                   title="relative growth")
```
