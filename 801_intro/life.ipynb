```python
country_l = list(df['name'])
countries_life_dict = {}

for n in country_l:
    countries_life_dict[n] = [float(i) for i in list(df[df['name'] == n].iloc[0])[2::]]

print(len(countries_life_dict))
```

```python
years = list(df.columns)[2::]
```

```python
import matplotlib.pyplot as plt

for country, values in countries_life_dict.items():
    plt.plot(years, values, label=country)

plt.xlabel("Year")
plt.ylabel("Value")

plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.figure(figsize=(10000, 10000))

plt.show()
```
