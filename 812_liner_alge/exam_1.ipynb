```python
import sympy as sp

def solve_system_general(A_matrix, b_vector):
    A = sp.Matrix(A_matrix)
    b = sp.Matrix(b_vector)
    
    augmented = A.col_insert(A.shape[1], b)
    rref_matrix, pivot_cols = augmented.rref()
    
    print("RREF")
    sp.pprint(rref_matrix)
    print(f"\nPivot Columns (主元行): {pivot_cols}")

    try:
        sol, params = A.gauss_jordan_solve(b)
        
        if len(params) == 0:
            print("結果：唯一解 (Unique Solution)")
            sp.pprint(sol)
        else:
            print("結果：無限多解 (Infinite Solutions)")
            print(f"自由變數 (Free Variables): {params}")
            print("\n表達式 (Parametric Form):")
            sp.pprint(sol)
            
    except ValueError:
        print("結果：無解 (Inconsistent System)")
        print("最後一列出現了 [0 0 ... 0 | 非0] 的情況。")


A_test = [
    [2,1,1],
    [1,0,1],
    [-1,1,-2]
]
b_test = [0,1,-3]

solve_system_general(A_test, b_test)
```
