```python
import numpy as np
v1 = np.array([-1,0,0])
w1 = np.array([1,0,0])

A = np.array([[1,0,0],
              [0,-1,0],
              [0, 0,-1]])

def ang(v,w):
    dot = v @ w
    angle = np.arccos(dot / (np.linalg.norm(v)*np.linalg.norm(w)))
    orthogonal = np.isclose(dot, 0)

    k = angle / np.pi
    frac = Fraction(k).limit_denominator(12)

    print("dot:\n",dot,"\northogonal:\n",orthogonal)
    print("angle (rad):", angle)
    print("angle ≈", frac, "π")
    

ang(v1,w1)
```
